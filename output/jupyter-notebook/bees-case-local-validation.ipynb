## Step 1 - Rode o pipeline local em PySpark

A proxima celula cria um `SparkSession`, usa o arquivo `examples/sample_breweries.json` e gera as camadas do projeto em `local_output/`.


```python
from bees_case.pyspark_local import create_spark_session, run_local_pyspark_pipeline

spark = create_spark_session(app_name="bees-case-colab")
summary = run_local_pyspark_pipeline(
    spark=spark,
    source_path="examples/sample_breweries.json",
    output_root="local_output",
    landing_date="2026-03-16",
    run_id="colab-pyspark-001",
)
summary
```

## Step 2 - Veja o conteudo gerado

Aqui a ideia e confirmar que as camadas `bronze`, `silver`, `gold` e `ops` foram produzidas.


```python
spark.read.json("local_output/bronze/landing_date=2026-03-16").show(truncate=False)
spark.read.parquet("local_output/silver/breweries").show(truncate=False)
spark.read.parquet("local_output/gold/breweries_by_type_location").show(truncate=False)
spark.read.parquet("local_output/ops/quality_results").show(truncate=False)
```
